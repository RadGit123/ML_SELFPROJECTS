```python
import zipfile
import pandas as pd

zip_path = r'C:\Users\SR\Desktop\entry+assignments\ML Assignments(notebook)/adult.zip'
csv_file = 'adult.data'

# Column names for Adult dataset (from 'adult.names' file)
columns = [
    "age", "workclass", "fnlwgt", "education", "education-num",
    "marital-status", "occupation", "relationship", "race", "sex",
    "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"
]

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    with zip_ref.open(csv_file) as file:
        df = pd.read_csv(file, header=None, names=columns, skipinitialspace=True)

print(df.head())
```
