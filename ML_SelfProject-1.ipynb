# *Cramer’s V to see all imortant catagorical features*.

```python
import pandas as pd
import numpy as np
import scipy.stats as st

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = st.chi2_contingency(confusion_matrix, correction=False)[0]
    n = confusion_matrix.sum().sum()
    k = min(confusion_matrix.shape) - 1
    return np.sqrt(chi2 / (n * k))

cat_cols = ['workclass','education','marital-status','occupation','relationship','race','sex','native-country']

print("\nCramer’s V (Overall association with target)")
for col in cat_cols:
    v = cramers_v(df[col], df['target'])
    print(f"{col.ljust(16)} → Cramér’s V = {v:.3f}")

print("\nCategory-wise Income Proportions.")
for col in cat_cols:
    print(f"\n{col.upper()}")
    income_rate = df.groupby(col)['target'].mean().sort_values(ascending=False)
    print(income_rate)
```

# *6.Outiner detection and removal*.

```python
import numpy as np
df_ml = Final_df.copy()
IQR_cols = ['age', 'education-num', 'hours-per-week']
outlier_indices = set()

for col in IQR_cols:
    Q1 = df_ml[col].quantile(0.25)
    Q3 = df_ml[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 3 * IQR
    upper = Q3 + 3 * IQR
    outliers = df_ml[(df_ml[col] < lower) | (df_ml[col] > upper)]
    outlier_indices.update(outliers.index)

df_ml = df_ml.drop(index=outlier_indices)
df_ml['capital-gain'] = np.log1p(df_ml['capital-gain'])                # beause most of the values are close to 0 so betetr use log transformaton thn IQR
df_ml['capital-loss'] = np.log1p(df_ml['capital-loss'])
```

# *Precision-Recall Curve*.

PRC focuses on precision (how many predicted positives are actually correct) and recall (how many actual positives were detected).
Since my dataset is imbalanced with only 25% of samples belonging to the '>50K' class, I used the Precision-Recall Curve to better evaluate how well my model captures the minority class, which is more informative than the ROC curve in this case."

```python
from sklearn.metrics import precision_recall_curve, auc
import matplotlib.pyplot as plt

y_prob_HPlog = L_Model.predict_proba(X_test_final)[:, 1]
precision, recall, thresholds = precision_recall_curve(y_test, y_prob_HPlog)
prc_auc = auc(recall, precision)

fig, ax = plt.subplots(figsize=(4, 4))
fig.patch.set_facecolor('#0d1a2b')   
ax.set_facecolor('#0d1a2b')          
ax.plot(recall, precision, label=f'Logistic Regression (AUC = {prc_auc:.3f})', color='cyan')
ax.set_xlabel('Recall', color='white')
ax.set_ylabel('Precision', color='white')
ax.set_title('Precision-Recall Curve', color='white')
ax.tick_params(axis='both', colors='white') 

ax.legend(facecolor='#0d1a2b', edgecolor='white', labelcolor='white')
ax.grid(True, color='gray', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

The curve starts near 1.0 precision at low recall > the model is very confident in the few positives it first detects.
As recall increases (i.e., catching more >50K earners), precision slowly drops, as expected ,more positive predictions include some false positives.
AUC = 0.749 is good for imbalanced binary classification
